```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##source https://pythonguides.com/matplotlib-change-background-color/
##https://pythonguides.com/matplotlib-set_xticklabels/
plt.rcdefaults()
plt.rcParams.update({'axes.facecolor':'black'})
plt.figure(facecolor='black') 

```

```python
df = pd.read_csv('Hoteldata.csv')
df.head(2)
```

```python
# df.info()
```

```python
##https://matplotlib.org/stable/gallery/color/named_colors.html
plt.figure(facecolor='grey') 
df['Neighbourhoods'].value_counts().plot(kind = 'bar',color='cornflowerblue' )
# set_xticklabels(df['Neighbourhoods'], color='red', fontsize=15)
# plt.xticks(df['Neighbourhoods'], color='red', fontsize=15)
# fig,ax = plt.subplots()
plt.xlabel('Locations')
plt.ylabel('Counts')
plt.title('Location where hotels are most dominated')
```

```python
plt.figure(facecolor='grey') 
df['Neighbourhoods'].value_counts().nlargest(10).plot(kind = 'pie',figsize=(8,8), autopct = '%1.1f%%')
plt.gca().axes.get_xaxis().set_visible(False)
plt.title('Cluster location of hotels')
```

```python
new= df[df['Ratings']>=3.5]
new = new.sort_values('Ratings',ascending=False)
```

```python
plt.figure(facecolor='grey') 
new[['Neighbourhoods','Ratings']].groupby('Ratings').value_counts().plot(kind='bar', figsize=(12,5))
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel('Counts')
plt.title('Rating based on Neighbourhoods')
```

```python
plt.figure(facecolor='grey') 
df[['Neighbourhoods','Number_of_reviews']].groupby('Neighbourhoods').sum().sort_values('Number_of_reviews',ascending=False).head(10).plot(kind='bar')
plt.ylabel('Counts')
plt.title('Neighbourhoods with most popular lodging')
```

```python
plt.figure(facecolor='grey') 
df[['Ratings','Prices']].groupby(['Prices']).value_counts().plot(kind='bar')
plt.ylabel('Counts')
plt.title('Most prefered price range')
```

```python
plt.figure(facecolor='grey') 
df[['Prices','Number_of_reviews']].groupby(['Prices']).sum().plot(kind='bar')
plt.ylabel('Counts')
plt.title('Most prefered price range')
```

```python
df2 = df[['Type_of_hotels','Number_of_reviews']]
df_sorted =df2.groupby('Type_of_hotels',as_index=False).sum().sort_values('Number_of_reviews',ascending=False).head(10)
# trying.sort_values('Number_of_reviews',ascending=False) 
# trying.head(10).plot(kind='bar')
df_sorted
```

```python
plt.figure(facecolor='grey') 
plt.barh(df_sorted['Type_of_hotels'],df_sorted['Number_of_reviews'])
plt.ylabel('Type of lodge')
plt.xlabel('Counts')
plt.title('Most visted lodge (B&Bs or hotels')
```
